1. What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?

**solution:**

Difference Between a Basic Filter, Advanced Filter, and a Slicer

**Basic Filter**: * Functionality: Operates within the Filters Pane. It allows users to select specific values from a list of checkboxes corresponding to a data column.  

User Interface: Hidden or collapsed on the side of the report page.

 Use Case: Ideal for simple inclusions/exclusions (e.g., ticking off specific years like 2023).

**Advanced Filter:** Functionality: Operates within the Filters Pane  but utilizes logical conditions rather than simple checkboxes. It allows for complex criteria like "Contains", "Starts with", "Is greater than", or combining conditions with AND/OR logic.  

 User Interface: Located within the Filters Pane, requiring text or numeric value inputs.

 Use Case: Best for pattern matching or range filtering (e.g., finding products with names containing "Smart" and sales greater than ₹10,000).

 **Slicer**: * Functionality: An on-canvas interactive visual element that acts as an alternative way of filtering data.  

 User Interface: Placed directly on the report canvas, highly customizable, and visible to the end-user as buttons, dropdowns, lists, or sliders.

  Use Case: Designed for executive dashboards where end-users need immediate, intuitive, and interactive filtering control without digging into the side panels.

4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?

**solution:**

A Visual-Level Filter is a specific type of filter in Power BI that applies data restrictions exclusively to a single visualization element on your report canvas. It allows you to isolate or clean up the data within one specific chart or table without impacting any other visuals around it.

 Power BI organizes filters hierarchically to control the scope of data isolation, and it differs from Page-Level and Report-Level filters in the following ways:

  1. Scope of Impact
  
  Visual-Level Filter: Affects only the single, specific visualization to which it is applied. If you have two bar charts side-by-side, applying a visual-level filter to one leaves the other completely untouched.
  
   Page-Level Filter: Affects all visualizations residing on that specific report page.  
   
   Report-Level Filter: Affects every single visualization across all pages of the entire Power BI report file.

6. What is the role of Drill-Down in Power BI? How does it improve data exploration?

**solution:**

The Drill-Down feature allows users to navigate from a high-level summarized data view down to a more granular, detailed level within the same visualization.

 **How it Improves Data Exploration:**  

Maintains Context: Instead of jumping to a completely different report page, the user stays within the same chart, preserving their mental model of the data.

Reduces Visual Clutter: Instead of showing messy, overwhelming charts containing hundreds of micro-level bars, it presents a clean high-level summary that users can selectively unpack.

Discovers Hidden Trends: It allows users to pinpoint exactly which lower-level component is driving a spike or anomaly visible at the summary level.

10. Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example.

**solution:**

the core structural difference lies in whether you are changing granularity within the same visual or moving to a different contextual sheet.

 Real-World Example: Retail Pharmacy Chain Analysis

 Drill-Down (Changing Granularity): You are looking at a single Bar Chart showing the total annual sales of the entire company. You click on the year bar, and it breaks down into 4 Quarter bars within that same chart. You click a Quarter, and it splits into Months. You stay on the same screen; the data simply changes focus.

  Drill-Through (Contextual Jumping): You are looking at an executive summary page showing performance across 50 different retail store branches. You spot a specific store that has a massive drop in revenue. You right-click that specific store name, choose "Drill-Through", and Power BI launches an entirely separate page dedicated to "Store Diagnostics".
  
   This new page shows completely different visuals like staff shift hours, local inventory levels, and customer complaints specific only to that store branch.

13. What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction.

**solution:**

When turned ON (Default): If your master report page has other active selections—such as a slicer set to "Year = 2023" and a filter set to "Segment = Online"—those constraints are carried along when you drill-through on a product. The detailed target page will inherit all of those filters simultaneously.

 When turned OFF: The target page will completely ignore your master page slicers. It will only filter itself by the exact data point you right-clicked (e.g., the Product name), showing global historic performance across all years and segments for that item, ignoring the rest of your current user session's state.
